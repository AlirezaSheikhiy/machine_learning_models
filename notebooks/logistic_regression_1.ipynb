### Import required modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
```

### Read dataset

```python
df = pd.read_csv("../datasets/titanic.csv")
```

### Understanding dataset

```python
df.head()
```

```python
print("Number of survived passengers:", df[df["Survived"] == 1].shape[0])
print("Number of not survived passengers:", df[df["Survived"] == 0].shape[0])
```

```python
sns.pairplot(df)
plt.show()
```

### Preprocessing

```python
df["male"] = df["Sex"] == "male"
df["male"] = df["male"].astype("int")
```

### Features and target selection

```python
X = df.drop(columns=["Survived", "Sex"], axis=1).values
y = df["Survived"].values
```

### Dataset train-test splitting

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=24)
```

### Regression

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

### Evaluation

```python
model.score(X_test, y_test)
```

### Prediction

Will a __50 years__ old __male__ passenger from __Pclass 1__ with __1 Spouse__ and __2 Children__ and __Fair of 30__, __Survive__?

```python
res = model.predict([[1, 50, 1, 2, 30, 1]])[0]
print("No") if res == 0 else print("Yes")
```
