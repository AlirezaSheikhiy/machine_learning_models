<h1 style="text-align:center;margin:50px 0">Polynomial Regression</h1>

### Import required modules

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

## 1. Linear Data

#### Generating random data

```python
X = np.random.rand(100, 1)
noise = np.random.randn(100, 1)

y1 = 5 * X + 4
y2 = 5 * X + 4 + 0.2 * noise    # low noise
y3 = 5 * X + 4 + noise    # normal noise
y4 = 5 * X + 4 + 7 * noise    # high noise
```

#### Plotting

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

axes[0,0].scatter(X, y1)
axes[0,0].set_title("No noise")
axes[0,0].set_xlabel("x")
axes[0,0].set_ylabel("y")

axes[0,1].scatter(X, y2)
axes[0,1].set_title("Low noise")
axes[0,1].set_xlabel("x")
axes[0,1].set_ylabel("y")

axes[1,0].scatter(X, y3)
axes[1,0].set_title("Normal noise")
axes[1,0].set_xlabel("x")
axes[1,0].set_ylabel("y")

axes[1,1].scatter(X, y4)
axes[1,1].set_title("High noise")
axes[1,1].set_xlabel("x")
axes[1,1].set_ylabel("y")

plt.tight_layout()

plt.show()
```

<p style="color:red">We can use <b>Linear Regression</b> for this types of linear data even with noise.</p>

## 2. Non-linear Data

```python
X = 4 * np.random.rand(100, 1) - 2
y = 2*X**3 + 2*X**2 + 2*X + 3 + 2 * noise
```

```python
plt.scatter(X, y)
plt.title("No noise")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

<p style="color:red"><b>Linear Regression</b> is not able to handle this type of data. We have to use <b>Polynomial Regression</b> for this data even with noise (Polynomial Curve Fitting).</p>

## 3. Regressions

#### Polynomial Regression (Underfit) 

```python
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
X_vals = np.linspace(-3, 3, 100).reshape(-1, 1)
X_vals_poly = poly_features.fit_transform(X_vals)
y_vals = model.predict(X_vals_poly)
```

```python
plt.scatter(X, y)
plt.plot(X_vals, y_vals, color="r")
plt.show()
```

#### Polynomial Regression (Overfit) 

```python
poly_features = PolynomialFeatures(degree=5, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
X_vals = np.linspace(-3, 3, 100).reshape(-1, 1)
X_vals_poly = poly_features.fit_transform(X_vals)
y_vals = model.predict(X_vals_poly)
```

```python
plt.scatter(X, y)
plt.plot(X_vals, y_vals, color="r")
plt.show()
```

#### Polynomial Regression (Fit) 

```python
poly_features = PolynomialFeatures(degree=3, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
X_vals = np.linspace(-2, 2, 100).reshape(-1,1)
X_vals_poly = poly_features.fit_transform(X_vals)
y_vals = model.predict(X_vals_poly)
```

```python
plt.scatter(X, y)
plt.plot(X_vals, y_vals, color="r")
plt.show()
```

<p style="color:green">Selectiong the best <b>polynomial degree</b> will cause your model fits the datas.</p>
